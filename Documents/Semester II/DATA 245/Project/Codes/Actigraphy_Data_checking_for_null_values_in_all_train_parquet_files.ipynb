```python
import os
import pyarrow.parquet as pq

# Defining the directory where the train parquet files are stored
parquet_dir = '/Users/nandana_hemanth/Documents/Semester II/DATA 245/Project/Dataset/child-mind-institute-problematic-internet-use/series_train.parquet' 
# Function to check missing values in a single parquet file
def check_missing_values(file_path):
    df = pq.read_table(file_path).to_pandas()
    missing_values = df.isnull().sum()  # Counting the missing values per column
    total_missing = missing_values.sum()  # Sum of all the missing values in the file
    return total_missing, missing_values

# Initializing counters and lists to store the results
missing_data_report = []
total_files = 0
files_with_missing = 0

# Iterating through the directory structure and checking all the parquet files for null values
for root, dirs, files in os.walk(parquet_dir):
    print(f"Current directory: {root}")  
    print(f"Files in this directory: {files}")  
    for file in files:
        if file == 'part-0.parquet':  
            total_files += 1
            file_path = os.path.join(root, file)
            
            # Extracting the directory name (id=xxxxxx) from the path
            relative_dir = os.path.relpath(root, parquet_dir) 
            file_identifier = f"id={relative_dir}/{file}" 
            
            print(f"Checking file: {file_identifier}")  

            # Checking for missing values
            total_missing, missing_values = check_missing_values(file_path)
            
            # If there are missing values, we will store the report
            if total_missing > 0:
                print(f"File {file_identifier} has missing values.")  
                files_with_missing += 1
                missing_data_report.append({
                    'file': file_identifier,
                    'total_missing': total_missing,
                    'missing_per_column': missing_values
                })

# Printing the summary report of missing data (i.e., only files with missing values will be printed)
if missing_data_report:
    for report in missing_data_report:
        print(f"File: {report['file']}")
        print(f"Total missing values: {report['total_missing']}")
        print(f"Missing values per column:\n{report['missing_per_column']}")
        print("--------------------------------------------------")
else:
    print("No files with missing values found.")  

print(f"Total files iterated through: {total_files}")
print(f"Files with missing values: {files_with_missing}")
```
