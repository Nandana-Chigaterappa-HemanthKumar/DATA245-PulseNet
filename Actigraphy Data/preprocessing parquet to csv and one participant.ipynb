```python
import dask.dataframe as dd
import os
import pandas as pd

# Path to the directory containing the Parquet files
parquet_dir = "C:/Users/Lenovo/Downloads/child-mind-institute-problematic-internet-use/series_train.parquet/"

# Output directory for CSV files
output_dir = "C:/Users/Lenovo/Downloads/child-mind-institute-problematic-internet-use/output_csv/"
os.makedirs(output_dir, exist_ok=True)  # Create the output directory if it doesn't exist

# Get a list of all Parquet files in the directory (including subdirectories)
file_paths = []
for root, dirs, files in os.walk(parquet_dir):
    for file in files:
        if file.endswith('.parquet'):
            file_paths.append(os.path.join(root, file))

# Check if any files were found
if not file_paths:
    raise ValueError("No Parquet files found in the specified directory.")

# Load each file, print column names, and save the first 5 rows to a CSV
for file_path in file_paths:
    print(f"\nProcessing file: {file_path}")
    
    # Load the Parquet file as a Dask DataFrame
    df = dd.read_parquet(file_path)
    
    # Fetch the column names and first 5 rows
    column_names = df.columns.tolist()
    first_5_rows = df.head(5)
    
    # Convert to a Pandas DataFrame with column names and first 5 rows
    output_df = pd.DataFrame(first_5_rows, columns=column_names)
    
    # Generate output CSV file path based on the Parquet file name
    output_file_name = os.path.splitext(os.path.basename(file_path))[0] + "_first_5_rows.csv"
    output_file_path = os.path.join(output_dir, output_file_name)
    
    # Save the output DataFrame to CSV
    output_df.to_csv(output_file_path, index=False)
    
    print(f"Saved to {output_file_path}")
```
