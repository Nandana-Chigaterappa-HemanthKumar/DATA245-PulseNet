```python
import os
import pandas as pd

# Directory containing the parquet files
input_directory = "C:/Users/Lenovo/OneDrive/Desktop/SEM 2/ml/project/cleaned_series_train.parquet"

# Initialized a set to store unique values across all files
unique_values = set()

# Listed all parquet files in the directory
parquet_files = [f for f in os.listdir(input_directory) if f.endswith('.parquet')]

# Processed each parquet file
for file in parquet_files:
    file_path = os.path.join(input_directory, file)
    try:
        # Read the parquet file
        data = pd.read_parquet(file_path)
        
        # Checked if 'non_wear_flag' column exists
        if 'non_wear_flag' in data.columns:
            # Add unique values from this file to the set
            unique_values.update(data['non_wear_flag'].unique())
        
        print(f"Processed file: {file}")
    except Exception as e:
        print(f"Error processing file {file}: {e}")

# Displayed the unique values found in the 'non_wear_flag' column
if unique_values:
    print("Unique values in 'non_wear_flag':", unique_values)
else:
    print("No 'non_wear_flag' column found in the files.")
```
